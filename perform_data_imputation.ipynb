```python
cols_with_nans = []
for i in data.columns:
    if (data[i].isnull().sum() != 0):
        cols_with_nans.append((i, data[i].isnull().sum()))

cols_with_nans
```

```python
data_dropped = data[data['deaths'].notna()]
```

```python
cols_with_nans2 = []
for i in data_dropped.columns:
    if (data_dropped[i].isnull().sum() != 0):
        cols_with_nans2.append((i, data_dropped[i].isnull().sum()))
```

### Perform Data Imputation Using KNN

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(n_neighbors=5, missing_values=np.nan, add_indicator=True)
after_imputing = knn.fit_transform(data_dropped)
```

```python
np.save('../data/imputed_data.npy', after_imputing)
```

```python
df_after_imputing = pd.DataFrame(after_imputing[:,0:194], columns = knn.get_feature_names_out()[0:194])
```

```python
df_after_imputing
```

### Save Imputed Data to CSV

```python
df_after_imputing.to_csv("../data/imputed_data.csv", index=False)
```
