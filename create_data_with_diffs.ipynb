```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/tableau_data.csv")
```

```python
print(data.shape)
data.head(15)
```

```python
data['datetime'] = pd.to_datetime(data['date'])
grouped = data.groupby(['state', 'county'])

final_dfs = []
keys = list(grouped.groups.keys())
for k in keys:
    curr = grouped.get_group(k).sort_values(by=['datetime'])
    curr['case_diffs'] = curr['cases'].diff()
    curr['death_diffs'] = curr['deaths'].diff()
    curr.case_diffs.fillna(curr.cases, inplace=True)
    curr.death_diffs.fillna(curr.deaths, inplace=True)
    final_dfs.append(curr)
```

```python
finalDf = pd.concat(final_dfs)
finalDf = finalDf.drop(columns=['datetime'])
print(finalDf.shape)
print(data.shape)
finalDf.head(50)
```

```python
finalDf['case_diffs'] = np.where(finalDf['case_diffs'] < 0, 0, finalDf['case_diffs'])
finalDf['death_diffs'] = np.where(finalDf['death_diffs'] < 0, 0, finalDf['death_diffs'])
finalDf.head(50)
```

```python
finalDf.to_csv("../data/tableau_data_with_diffs.csv", sep=',', index=False)
```
